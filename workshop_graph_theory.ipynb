```python
class Graph(object):
      
    def __init__(self,gr_dict):#it is the constructor of the class,and it is called when we create an object of this class
        if gr_dict is None:
            gr_dict={}
        self.graph_dict=gr_dict       
        
    def vertices(self):#return a list of vertices which are the keys of the dictionary self.graph_dict
         return list(self.graph_dict.keys()) 
    
    def edges(self):#return a list of sets, when a set represents an edge
        edges = []
        for key in self.graph_dict.keys():
            for vertex in self.graph_dict[key]:
                if {key,vertex} not in edges:
                    edges.append({key, vertex})
        return edges
       
    def addVertex(self,vertex):# add a vertex only if it doesn't exist
        if vertex not in self.graph_dict.keys():
            self.graph_dict[vertex]=[]
       
    def addEdge(self,tuple_edge):# add an edge even if a node doesn't exist
        (vertex1,vertex2)=tuple_edge
        if vertex1 not in self.graph_dict.keys():
            self.graph_dict[vertex1]=[vertex2]
        else:
            self.graph_dict[vertex1].append(vertex2) 
        if vertex2 not in self.graph_dict.keys():
            self.graph_dict[vertex2]=[vertex1]
        else:
            self.graph_dict[vertex2].append(vertex1) 
        
    def isolated_Nodes(self):
        isolatedNodes = []
        for node in self.graph_dict.keys():
            if not self.graph_dict[node]:
                isolatedNodes.append(node)
        return isolatedNodes
    
    def find_a_path(self, start_vertex, end_vertex, path=[]):
        isolatedNodes=self.isolated_Nodes()
        path = path + [start_vertex]
        if start_vertex == end_vertex:
            return path
        if start_vertex in isolatedNodes or end_vertex in isolatedNodes:
            return None
        for vertex in self.graph_dict[start_vertex]:
            if vertex not in path:
                extended_path = self.find_a_path(vertex,end_vertex,path)
                if extended_path:
                    return extended_path
        return None
    
    def find_all_paths(self, start_vertex, end_vertex, path=[]):
        isolatedNodes=self.isolated_Nodes()
        path = path + [start_vertex]
        if start_vertex == end_vertex:
            return [path]
        if start_vertex not in self.graph_dict or start_vertex in isolatedNodes or end_vertex in isolatedNodes:
            return None
        paths=[]
        for vertex in self.graph_dict[start_vertex]:
            if vertex not in path:
                extended_paths = self.find_all_paths(vertex,end_vertex,path)
                if extended_paths is not None:
                    for p in extended_paths:
                        paths.append(p)
        return paths
     
    def minPath(self, start_vertex, end_vertex):
        paths=self.find_all_paths(start_vertex, end_vertex, path=[])
        if paths is not None:
            smallest = sorted(paths, key=len)[0]
            return smallest
        else: 
            return
    
    def degreeOfVertex(self,vertex):
        return #code here 
    
    def density(self):
        return #code here
    
    def is_connected(self):
        return #code here
    
    def diameter(self):
        v = self.vertices()
        pairs = [(v[i], v[j]) for i in range(len(v) - 1) for j in range(i + 1, len(v))]#all possible Pairs
        smallest_paths = []
        for (s, e) in pairs:
            smallest = self.minPath(s, e)#minimum path for each pair
            if smallest is not None:
                smallest_paths.append(smallest)
        smallest_paths.sort(key=len)# to find the maximum path between all shortest parths
        diameter = len(smallest_paths[-1]) - 1
        return diameter
```
